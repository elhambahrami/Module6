#### In this jupyter notebook I will work with a fixed file that includes precipitation measurments inorder to ingest it.
#### The file is already downloaded in text file.

```python
import pandas as pd
```

#### Since I need to read the file I nead pandas read_fwf(for fixed width files) and also to skip some rows after exploratoring the file to figure out how many rows I should skip.

```python
prep_df = pd.read_fwf('/Users/elham/Documents/mctyprcp.webarchive', 
                        skiprows= 11,
                        header = None)
prep_df.head()
```

#### Let's look at the last rows of the data

```python
prep_df.tail()
```

#### Looks like I some problems with the last row and should remove it.

```python
prep_df =prep_df.drop(251)
```

```python
prep_df.tail()
```

#### Now looks better

#### As I look at the values in the columns I realized that column '1' has two values in, so it has to be splited.

```python
prep_df['city'] = prep_df[1].str[:-1]
prep_df['pflag'] = prep_df[1].str[-1]
```

```python
prep_df.head()
```

#### So we don't need column '1' any more.

```python
prep_df = prep_df.drop(columns=1)
```

```python
prep_df.head()
```

#### So I need to do the same thing with column 10 which has two values.

```python
prep_df['year_to_date_amt'] = prep_df[10].str[:4]
prep_df['year_to_date_dep_nml'] = prep_df[10].str[-5:]
```

```python
prep_df.head()
```

#### Droping column 10

```python
prep_df = prep_df.drop(columns=10)
prep_df.head()
```

#### Now it is the time to fix the columns name

```python
prep_df.rename(columns={0:'state',
                          2: "call_ltrs",
                          3: "month_tot_amt",
                          4: "month_dep_nml",
                          5: "24_hr_max",
                          6: "days_gt_1",
                          7: "days_gt_50",
                          8: "season_to_date",
                          9: "season_dep_nml",
                          11: "year_to_date_pct_nml"},inplace=True)
```

```python
prep_df.head()
```

#### Reordering the columns

```python
prep_df = prep_df[["state",
                       "city",
                       "pflag","call_ltrs","month_tot_amt",
                   "month_dep_nml","24_hr_max","days_gt_1","days_gt_50","season_to_date",
                   "season_dep_nml","year_to_date_pct_nml","year_to_date_amt","year_to_date_dep_nml"]]
prep_df.head()
```

#### Let's see howmany null values we got

```python
prep_df.isnull().sum()
```

#### So since it is NOT that many it is better to drop them off.

```python
prep_df.dropna(inplace=True)
```

```python
prep_df.isnull().sum()
```

#### Now I want to figure out which city has the highest year_to_date_pct_nml

```python
prep_df = prep_df.sort_values(by=['year_to_date_pct_nml'])
```

```python
#lowest YTD-pct normal
prep_df.iloc[1]
```

```python
#highest YTD pct normal, there is name of city tho!!!
prep_df.iloc[248]
```

#### Now it is the time to put the data in a neated file!
